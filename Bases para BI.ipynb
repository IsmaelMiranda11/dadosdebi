```python
##Lista de consumo enegértico

num_linhas = 35000


#Segmento
segmento = ['SUV','Compacto','Picape','Luxo','Minivan','Sedã']
seg_prob = [0.25, 0.4, 0.15, 0.05, 0.1, 0.05]
producaomed = [50, 80, 30, 10, 20, 10]
segs = np.random.choice(segmento, num_linhas, p = seg_prob)

#Setor
setores= ['Estamparia','Estruturação','Funilaria','Pintura', 'Portas', 'Motor', 'Montagem']
prop_set = [0.6, 0.1, 0.1, 0.05, 0.05, 0.05, 0.05]
consumo_med = [600.0000, 251.3443, 800.5234, 102.2368, 24.3222, 154.3224, 652.4597]

seg_setor = np.transpose([np.tile(segs, len(setores)), np.repeat(setores, len(segs))])
num_linhas = len(seg_setor)

#Produto, emcima do segmento
def produto(df): #Recebe o dataset como entrada
    nm_seg = df['Segmento']
    nm_prod = nm_seg[0:3] + " "  "%02d" % np.random.choice(range(1,10),1)
    return nm_prod

#Produção
def producao(df): #Recebe o dataset como entrada
    ind_ = segmento.index(df['Segmento']) #Verifica o cargo da linha
    prod_ = producaomed[ind_] #Usa a lista auxiliar
    ind2_ = setores.index(df['Setor'])
    prop_ = prop_set[ind2_]
    mean = prod_*prop_; stdev = mean*0.02
    return mean + np.random.randn() * stdev

oi = pd.date_range(end='1/1/2018', periods=8)
print(oi)

def consumo(df): #Recebe o dataset como entrada
    ind_ = setores.index(df['Setor']) #Verifica o cargo da linha
    cons_ = consumo_med[ind_] #Usa a lista auxiliar
    mean = cons_; stdev = mean*0.02
    return mean + np.random.randn() * stdev


#Datas
datas = np.arange('2015-01', '2020-06', dtype='datetime64[D]')
datas = np.random.choice(datas, num_linhas)

prod = pd.DataFrame({'Segmento':seg_setor[:,0], 'Setor': seg_setor[:,1], 'Data': datas})
prod['Produto'] = prod.apply(produto,axis=1)
prod['Producao'] = prod.apply(producao,axis=1)

prod.to_excel('prod.xlsx')

cons = pd.DataFrame({'Segmento':seg_setor[:,0], 'Setor': seg_setor[:,1], 'Data': datas})
cons['Produto'] = prod.apply(produto,axis=1)
cons['Consumo kWh'] = prod.apply(consumo,axis=1)

cons.to_excel('cons.xlsx')
```
